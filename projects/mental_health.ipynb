```python
"""
correlation_pca_acf_analysis.py

Performs:
 - correlation matrix + p-values (Pearson)
 - alternative rank correlations (Spearman, Kendall) summary
 - PCA (standardized): scree plot, explained variance, component loadings
 - Bartlett's test of sphericity and KMO measure (PCA suitability)
 - ACF for each numeric feature and Ljung-Box test for autocorrelation
 - Plots: correlation heatmap (p-value mask), scree plot, PCA loadings heatmap, ACF plots
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.stattools import acf
from statsmodels.stats.diagnostic import acorr_ljungbox
import itertools

# --------- Helpers ---------
def pairwise_pearson_pvals(df_numeric):
    """Return correlation matrix and matrix of p-values (Pearson)."""
    cols = df_numeric.columns
    n = len(cols)
    corr = pd.DataFrame(np.zeros((n, n)), index=cols, columns=cols)
    pvals = pd.DataFrame(np.zeros((n, n)), index=cols, columns=cols)
    for i, j in itertools.product(range(n), range(n)):
        if i == j:
            corr.iloc[i, j] = 1.0
            pvals.iloc[i, j] = 0.0
        else:
            r, p = stats.pearsonr(df_numeric.iloc[:, i], df_numeric.iloc[:, j])
            corr.iloc[i, j] = r
            pvals.iloc[i, j] = p
    return corr, pvals

def pairwise_rank_corrs(df_numeric):
    """Return Spearman and Kendall correlation dataframes."""
    spearman = df_numeric.corr(method='spearman')
    kendall = df_numeric.corr(method='kendall')
    return spearman, kendall

def bartlett_sphericity_test(corr_matrix, n_samples):
    """
    Bartlett's test of sphericity.
    Chi-square = -(n - 1 - (2p + 5)/6) * ln(det(R))
    df = p*(p-1)/2
    """
    p = corr_matrix.shape[0]
    det_r = np.linalg.det(corr_matrix)
    # guard against numerical issues
    if det_r <= 0:
        # if determinant non-positive, test is not defined properly
        return np.nan, np.nan
    chi2 = -(n_samples - 1 - (2*p + 5)/6.0) * np.log(det_r)
    df = p*(p-1)//2
    p_value = stats.chi2.sf(chi2, df)
    return chi2, p_value

def kmo_test(df_numeric):
    """
    Kaiser-Meyer-Olkin (KMO) measure.
    Implementation based on partial correlations:
        KMO = sum_{i != j} r_ij^2 / (sum_{i != j} r_ij^2 + sum_{i != j} p_ij^2)
    where p_ij are partial correlations.
    Returns overall_kmo, kmo_per_variable (Series).
    """
    corr = np.corrcoef(df_numeric.values.T)
    inv_corr = np.linalg.inv(corr)
    # compute partial correlation matrix
    # partial_ij = -inv_corr[i,j] / sqrt(inv_corr[i,i]*inv_corr[j,j])
    d = np.sqrt(np.diag(inv_corr))
    denom = np.outer(d, d)
    partial = -inv_corr / denom
    np.fill_diagonal(partial, 0.0)

    # sum of squares off-diagonal
    corr_sq = (corr - np.eye(corr.shape[0]))**2
    partial_sq = partial**2

    ss_corr = corr_sq.sum()
    ss_partial = partial_sq.sum()
    overall_kmo = ss_corr / (ss_corr + ss_partial)

    # per-variable KMO
    kmo_per_var = []
    p = corr.shape[0]
    for i in range(p):
        numer = corr_sq[i, :].sum() - corr_sq[i, i]
        denom_i = numer + (partial_sq[i, :].sum() - partial_sq[i, i])
        kmo_per_var.append(numer/denom_i if denom_i != 0 else np.nan)

    cols = df_numeric.columns
    kmo_series = pd.Series(kmo_per_var, index=cols)
    return overall_kmo, kmo_series

def make_corr_heatmap(corr, pvals=None, p_threshold=0.05, title="Correlation heatmap", figsize=(9,9), savepath=None):
    """Plot correlation heatmap using matplotlib; optionally mask non-significant cells by pvals."""
    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(corr, vmin=-1, vmax=1, interpolation='nearest')
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    cols = corr.columns
    ax.set_xticks(range(len(cols)))
    ax.set_xticklabels(cols, rotation=45, ha='right')
    ax.set_yticks(range(len(cols)))
    ax.set_yticklabels(cols)
    ax.set_title(title)

    # overlay p-values: put significance markers or mask
    if pvals is not None:
        for (i, j), val in np.ndenumerate(corr.values):
            p = pvals.values[i, j]
            text = f"{val:.2f}"
            if p > p_threshold and i != j:
                # mark as not significant with parentheses or lighter text
                ax.text(j, i, text, ha='center', va='center', color='lightgray', fontsize=8)
            else:
                ax.text(j, i, text, ha='center', va='center', color='black', fontsize=8)

    plt.tight_layout()
    if savepath:
        plt.savefig(savepath, dpi=150, bbox_inches='tight')
    plt.show()

# --------- Main analysis ---------
def run_analysis(csv_path="mental_health_social_media_dataset.csv", numeric_only=True):
    df = pd.read_csv(csv_path)
    # select numeric columns (if numeric_only) else try to coerce numeric ones
    if numeric_only:
        df_numeric = df.select_dtypes(include=[np.number]).copy()
    else:
        df_numeric = df.apply(pd.to_numeric, errors='coerce')

    # drop columns with too many NaNs
    df_numeric = df_numeric.dropna(axis=1, thresh=int(0.5*len(df_numeric)))
    df_numeric = df_numeric.dropna(axis=0)  # drop rows with any missing numeric values for simplicity
    n_samples = df_numeric.shape[0]
    print(f"Using {n_samples} samples and {df_numeric.shape[1]} numeric features for analysis.\n")
    print("Numeric features:", list(df_numeric.columns))

    # 1) Correlations + p-values (Pearson)
    corr, pvals = pairwise_pearson_pvals(df_numeric)
    print("\nPearson correlation matrix (head):")
    print(corr.iloc[:6, :6])
    print("\nPearson p-values matrix (head):")
    print(pvals.iloc[:6, :6])

    # plot correlation heatmap with significance annotation
    make_corr_heatmap(corr, pvals=pvals, p_threshold=0.05, title="Pearson correlation (values) - gray=not significant")

    # 2) Rank correlations
    spearman, kendall = pairwise_rank_corrs(df_numeric)
    print("\nTop absolute Pearson correlations (excluding self):")
    abscorr = corr.abs().unstack().sort_values(ascending=False)
    abscorr = abscorr[abscorr < 1]  # remove self correlations
    # drop duplicate pairs
    abscorr = abscorr.reset_index()
    abscorr.columns = ['var1', 'var2', 'abs_r']
    abscorr['pair'] = abscorr.apply(lambda row: tuple(sorted([row['var1'], row['var2']])), axis=1)
    abscorr = abscorr.drop_duplicates('pair').sort_values('abs_r', ascending=False)
    print(abscorr[['var1','var2','abs_r']].head(10))

    # print Spearman top pairs for comparison
    abs_spearman = spearman.abs().unstack().sort_values(ascending=False)
    abs_spearman = abs_spearman[abs_spearman < 1].reset_index()
    abs_spearman.columns = ['var1','var2','abs_r']
    abs_spearman['pair'] = abs_spearman.apply(lambda row: tuple(sorted([row['var1'], row['var2']])), axis=1)
    abs_spearman = abs_spearman.drop_duplicates('pair').sort_values('abs_r', ascending=False)
    print("\nTop absolute Spearman correlations:")
    print(abs_spearman[['var1','var2','abs_r']].head(10))

    # 3) PCA
    scaler = StandardScaler()
    X = scaler.fit_transform(df_numeric.values)
    pca = PCA(n_components=min(df_numeric.shape[1], 10))  # up to 10 components or number of features
    pcs = pca.fit_transform(X)
    explained = pca.explained_variance_ratio_
    cum_explained = np.cumsum(explained)
    print("\nPCA explained variance ratio (first components):")
    for i, ev in enumerate(explained, 1):
        print(f" PC{i}: {ev:.4f}  (cumulative {cum_explained[i-1]:.4f})")

    # Scree plot
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(1, len(explained)+1), explained, marker='o')
    plt.xlabel("Principal component")
    plt.ylabel("Explained variance ratio")
    plt.title("PCA Scree plot")
    plt.xticks(np.arange(1, len(explained)+1))
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # PCA loadings heatmap (components x features)
    loadings = pd.DataFrame(pca.components_.T,
                            index=df_numeric.columns,
                            columns=[f"PC{i+1}" for i in range(pca.n_components_)])
    plt.figure(figsize=(max(8, 0.6*len(df_numeric.columns)), 6))
    im = plt.imshow(loadings.T, aspect='auto', cmap='bwr', vmin=-1, vmax=1)
    plt.colorbar(im, fraction=0.05, pad=0.04)
    plt.yticks(range(loadings.shape[1]), loadings.columns)
    plt.xticks(range(len(loadings.index)), loadings.index, rotation=90)
    plt.title("PCA loadings (components x features)")
    plt.tight_layout()
    plt.show()
    print("\nPCA loadings (absolute sorted) for PC1:")
    print(loadings['PC1'].abs().sort_values(ascending=False).head(10))

    # 4) Bartlett and KMO
    corr_matrix = df_numeric.corr().values
    chi2, p_barlett = bartlett_sphericity_test(corr_matrix, n_samples)
    overall_kmo, kmo_per_var = kmo_test(df_numeric)
    print(f"\nBartlett's test chi2 = {chi2:.3f}, p-value = {p_barlett:.3e}")
    print(f"KMO overall = {overall_kmo:.3f}")
    print("KMO per variable (top 10):")
    print(kmo_per_var.sort_values(ascending=False).head(10))

    # 5) ACF and Ljung-Box per feature
    max_lag = min(30, int(0.2*n_samples))  # up to 30 or 20% of samples
    acf_results = {}
    lb_results = {}
    for col in df_numeric.columns:
        series = df_numeric[col].values
        # demean series to focus on autocorrelation
        series = series - np.mean(series)
        acf_vals = acf(series, nlags=max_lag, fft=True, missing='none')
        acf_results[col] = acf_vals
        # Ljung-Box test (returns test statistic and p-value arrays; use last lag)
        try:
            lb = acorr_ljungbox(series, lags=[min(10, max_lag)], return_df=True)
            lb_results[col] = lb.iloc[-1].to_dict()  # statistic and p-value
        except Exception as e:
            lb_results[col] = {'lb_stat': np.nan, 'lb_pvalue': np.nan}

    # plot ACFs: show first 8 features in a grid
    cols_to_plot = df_numeric.columns[:8]
    n_plots = len(cols_to_plot)
    ncols = 2
    nrows = int(np.ceil(n_plots / ncols))
    fig, axes = plt.subplots(nrows, ncols, figsize=(10, 3*nrows), squeeze=False)
    for idx, col in enumerate(cols_to_plot):
        r = idx // ncols
        c = idx % ncols
        axes[r, c].stem(np.arange(len(acf_results[col])), acf_results[col], basefmt=" ")
        axes[r, c].axhline(0, linestyle='--', color='gray')
        axes[r, c].set_title(f"ACF: {col}")
        axes[r, c].set_xlabel("Lag")
        axes[r, c].set_ylabel("ACF")
    plt.tight_layout()
    plt.show()

    # print Ljung-Box summary
    print("\nLjung-Box test summary (lag= up to min(10,max_lag)):")
    for col, res in lb_results.items():
        stat = res.get('lb_stat') if 'lb_stat' in res else (res.get('lb_statistic') if 'lb_statistic' in res else np.nan)
        p = res.get('lb_pvalue') if 'lb_pvalue' in res else (res.get('lb_pvalue') if 'lb_pvalue' in res else np.nan)
        print(f" {col}: stat={stat:.3f}, p={p:.3e}")

    # 6) Additional statistical tests: pairwise Pearson/Spearman/Kendall for top correlated pairs
    print("\nPairwise tests for top 10 absolute Pearson-correlated pairs (Pearson p and Spearman rho):")
    top_pairs = abscorr[['var1','var2']].head(10).values
    for v1, v2 in top_pairs:
        x = df_numeric[v1].values
        y = df_numeric[v2].values
        r, p = stats.pearsonr(x, y)
        rho, p_s = stats.spearmanr(x, y)
        tau, p_k = stats.kendalltau(x, y)
        print(f" {v1} - {v2} : Pearson r={r:.3f}, p={p:.3e} | Spearman rho={rho:.3f}, p={p_s:.3e} | Kendall tau={tau:.3f}, p={p_k:.3e}")

    # Save important objects to disk (optional)
    corr.to_csv("pearson_correlation_matrix.csv")
    pvals.to_csv("pearson_pvalues_matrix.csv")
    loadings.to_csv("pca_loadings.csv")
    print("\nSaved: pearson_correlation_matrix.csv, pearson_pvalues_matrix.csv, pca_loadings.csv")

if __name__ == "__main__":
    run_analysis("mental_health_social_media_dataset.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA

# Models
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report

# Load data
df = pd.read_csv("mental_health_social_media_dataset.csv")

# ================================
# 1. PREPROCESSING
# ================================

# Encode categorical features
categorical_cols = ["gender", "platform", "mental_state"]
encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

# Select predictors + target
# Target example: predict stress_level
target = "stress_level"
X = df.drop(columns=[target, "person_name", "date"])  # remove non-numeric + target
y = df[target]

# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ================================
# 2. PCA (Dimensionality Reduction)
# ================================

pca = PCA(n_components=0.95)   # keep 95% variance
X_pca = pca.fit_transform(X_scaled)

print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total components kept:", pca.n_components_)

# ================================
# 3. Train/Test Split
# ================================

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# ================================
# 4. PREDICTIVE MODELS
# ================================

# --- Model A: Linear Regression (baseline mathematical model) ---
linreg = LinearRegression()
linreg.fit(X_train, y_train)
y_pred_lin = linreg.predict(X_test)
mse_lin = mean_squared_error(y_test, y_pred_lin)

# --- Model B: Logistic Regression (if target is categorical) ---
# Example: predict "mental_state"
clf_log = LogisticRegression(max_iter=500)
clf_log.fit(X_train, y_train)
y_pred_log = clf_log.predict(X_test)

# --- Model C: Random Forest Regressor (non-linear model) ---
rf_reg = RandomForestRegressor(n_estimators=300, random_state=42)
rf_reg.fit(X_train, y_train)
y_pred_rf = rf_reg.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)

# ================================
# 5. RESULTS
# ================================

print("\n=== LINEAR REGRESSION (Regression) ===")
print("MSE:", mse_lin)

print("\n=== RANDOM FOREST REGRESSOR ===")
print("MSE:", mse_rf)

print("\n=== LOGISTIC REGRESSION (Classification) ===")
print(classification_report(y_test, y_pred_log))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# =========================================================
# 1. LOAD & PREPROCESS DATA
# =========================================================

df = pd.read_csv("mental_health_social_media_dataset.csv")

# Encode categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
encoders = {}

for col in categorical_cols:
    enc = LabelEncoder()
    df[col] = enc.fit_transform(df[col])
    encoders[col] = enc

# Choose target
target = "stress_level"

X = df.drop(columns=[target])
y = df[target]

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# =========================================================
# 2. TRAIN MODEL (Random Forest)
# =========================================================

model = RandomForestRegressor(n_estimators=300, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Performance
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)

# =========================================================
# 3. PLOT PREDICTION EVOLUTION
# =========================================================

plt.figure(figsize=(12, 6))
plt.plot(y_test.values, label="True Values", linewidth=2)
plt.plot(y_pred, label="Predicted Values", linestyle="--")

plt.title("Prediction Evolution Over Test Samples")
plt.xlabel("Sample Index")
plt.ylabel(target)
plt.legend()
plt.tight_layout()
plt.show()
```
